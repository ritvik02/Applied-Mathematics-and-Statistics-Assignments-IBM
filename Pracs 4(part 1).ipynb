```python
#frequency table
pd.crosstab(index=cars_data['FuelType'],columns='count',dropna=True)
```

```python
#2 way table
pd.crosstab(index=cars_data['Automatic'],columns=cars_data['FuelType'],normalize=True,dropna=True)
```

```python
#joint probability
pd.crosstab(index=cars_data['Automatic'],columns=cars_data['FuelType'],dropna=True)
```

```python
#marginal probability(it is prob of ocurrennce of a single event)
pd.crosstab(index=cars_data['Automatic'],columns=cars_data['FuelType'],margins=True,normalize=True,dropna=True)
```

```python
#conditonaly probability
pd.crosstab(index=cars_data['Automatic'],columns=cars_data['FuelType'],margins=True,normalize='index',dropna=True)
```
