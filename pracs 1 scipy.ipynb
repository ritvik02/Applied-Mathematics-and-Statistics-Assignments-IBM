```python
from pylab import*
#create some test data
dx=.001
x=np.arange(-2,2,dx)
y=exp(-x**2)
```

```python
#normalise the data to a proper PDF
y/=(dx*y).sum()

#compute the CDF
cy=np.cumsum(y*dx)
```

```python
#plot both
plot(x,y)
plot(x,cy,'r--')

plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)

plt.subplot(1,2,1)
plot(x,y)

plt.subplot(1,2,2)
plot(x,z)

plt.show()
```
