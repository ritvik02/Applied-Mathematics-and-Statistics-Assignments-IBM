```python
# 3.histogram
```

```python
plt.hist(df['Price'])
plt.xlabel('Price')
plt.ylabel('No of Cars')
plt.title('Histogram')
plt.show()
```

```python
# 8.distplot
```

```python
sns.distplot(df['Price'],kde=True).set_title('Histogram of Car price')
plt.show()      
```

```python
sns.distplot(df['Price'],kde=False).set_title('Histogram of Car price')
plt.show()   
```

```python
# 9.countplot
```

```python
y=df.FuelType
ax=sns.countplot(y,label='count')
petrol,diesel,cng=y.value_counts()
print('number of petrol cars:',petrol)
print('number of diesel cars:',diesel)
print('number of diesel cars:',cng)
```
