```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv('diabetes.csv')
df
```

```python
sns.lmplot(x='Age',y='Glucose',data=df)
plt.show()
```

```python
sns.lmplot(x='Age',y='Glucose',data=df,fit_reg=False #no regression line
           ,hue='Outcome') #color by evolution stage
plt.show()
```

```python
sns.swarmplot(x='Outcome',y='Glucose',data=df)
plt.show()
```

```python
#boxplot
sns.boxplot(data=df.loc[df['Outcome']==1,])
```
