```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_data=sns.load_dataset("tips")
```

```python
tips_data.head(20)
```

```python
tips_data.describe()
```

```python
#plot histogram of the total bill
sns.distplot(tips_data['total_bill'],kde=True).set_title('Histogram of total bill')
plt.show()                                   
#A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions. The approach is explained further in the user guide.
```

```python
#Histograms represent the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin. Seaborn comes with some datasets and we have used few datasets in our previous chapters.
```

```python
#plot histogram of the total bill
sns.distplot(tips_data['total_bill'],kde=False).set_title('Histogram of total bill')
plt.show()    
```

```python
sns.distplot(tips_data['tip'],kde=True).set_title('Histogram of total tip')
plt.show()
```

```python
sns.distplot(tips_data['total_bill'],kde=False)
sns.distplot(tips_data['tip'],kde=False).set_title('histo')
plt.show()
```

```python
sns.boxplot(tips_data['tip']).set_title('Box plot of the tip')
plt.show()
```

```python
sns.boxplot(tips_data['total_bill'])
sns.boxplot(tips_data['tip']).set_title('Box plot of the tip')
plt.show()
```

```python
sns.boxplot(x=tips_data['tip'],y=tips_data['time'])

g=sns.FacetGrid(tips_data,row='time')
g=g.map(plt.hist,'tip')
plt.show()
```

```python
sns.boxplot(x=tips_data['tip'],y=tips_data['time'])
```

```python
import seaborn as sns
g=sns.FacetGrid(tips_data,row='sex',col='smoker')
g=g.map(plt.hist,'tip')
```

```python
sns.boxplot(x=tips_data['tip'],y=tips_data['day'])

g=sns.FacetGrid(tips_data,row='day')
g=g.map(plt.hist,'tip')
plt.show()
```

```python
#bivariate plot.....scatte plots
sns.relplot(x='total_bill',y='tip',data=tips_data);
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=tips_data)
```

```python
#categorical data
sns.catplot(x='sex',y='tip',data=tips_data)
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='swarm',data=tips_data)
```

```python
sns.catplot(x='smoker',y='tip',order=['No','Yes'],
            data=tips_data)
```

```python
sns.catplot(x='smoker',y='tip',
            data=tips_data)
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='violin',data=tips_data)
```
